```python
import h5py

# Open the file
f = h5py.File('/home/kangrui/projects/world_model/diamond/data/untar_csgo/hdf5_dm_july2021_1.hdf5', 'r')

# List all groups
print("Groups in the file:")
print(list(f.keys()))

# Function to recursively explore groups and datasets
def explore_hdf5(item, indent=4):
    for key in item.keys():
        child = item[key]
        print(" " * indent + f"- {key}: ", end="")
        
        if isinstance(child, h5py.Group):
            print("Group")
            explore_hdf5(child, indent + 2)
        elif isinstance(child, h5py.Dataset):
            print(f"Dataset: shape {child.shape}, dtype {child.dtype}")
```
